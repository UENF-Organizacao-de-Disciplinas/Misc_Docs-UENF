```python
# Importações de bibliotecas

import cv2
from PIL import Image
import os
```

```python
# Brincadeira inicial


path = '/content/RGB.png'
with Image.open(path) as im:
  # im.show()
  bands = im.getbands()
  # print(bands)
  img_red = Image.new("L", im.size)
  img_gre = Image.new("L", im.size)
  img_blu = Image.new("L", im.size)
  img_alp = Image.new("L", im.size)

  r, g, b, a = im.split()
  img_red = Image.merge("RGB", (r, img_red, img_red))
  img_gre = Image.merge("RGB", (img_gre, g, img_gre))
  img_blu = Image.merge("RGB", (img_blu, img_blu, b))
  img_alp = Image.merge("RGB", (a, img_alp, img_alp))

  img_red.show()
  img_blu.show()
  img_gre.show()
  img_alp.show()
  im.show()

```
